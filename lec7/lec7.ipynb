**Bagging (Yeniden rneklem yntemi veya Torbalama)**

- Leo Breiman tarafndan nerilmitir (1996). Ama, snflandrma ve regresyon problemlerinde doruluunu arttrmak iin gelitirilmi bir tekniktir.

- Bagging (Bootstrapping Aggregating) metodu, var olan bir eitim setinden yeni eitim setleri treterek temel reniciyi yeniden eiten bir yntemdir. Her bir veri seti iin eitilen temel renicinin belirli bir test rnei zerindeki kararlar ortalama alnarak hesaplanr. 
- Baggingde ama yeni veri setleri treterek farkllklar oluturmak ve bu sayede toplam snflandrma baarsn artrmaktr. Bu yntemde eitim setinden yaklak olarak %63,2 kadar orijinal rnek **rastgele** alnr ve alnan rneklerden bazlar oaltlarak (resampling) eitim seti %100e tamamlanr. 
- Bu yntemle birbirinden farkl bir miktar eitim seti elde edilir. Her eitim seti ayn temel reniciye uygulanr ve alnan kararlar arlkl oylama yntemiyle birletirilir (Snflandrma iin). 
- Regresyon problemlerinde ise temel renicilerin kararlar ortalama ile belirlenir.

**Boosting**

- Boosting, bir ok zayf reniciyi bir araya getirerek bir gl renici (strong learner) oluturmak anlamna gelir. Bir ok boosting metodunun ana fikri, tahmin edicileri ardk olarak eitmektir. 

- Boosting yaps ise bamsz paralel yaplar yerine, srayla birbirinin uzerine eklenen yapya sahiptir. Bu tahmin yapma yntemleri, bir nceki tahminlerin hatalarndan renilen mant gelitirerek sonuca daha ok yaklamay hedefleyen modellerdir. Bu modellerde her seferinde bir ncekinin hatalarn gz nunde bulundurarak ilerledii iin paralel alan modellere gre daha hzl alrlar. Boosting yntemi, bagging ynteminden farkl olarak paralel karar aalarnn birletirilmesi yerine her uretilen karar aac, srayla hatay minimize edecek ekilde alr. Temelde, bu tip modellerde her yeni karar aac belirli arlk oranlarnda bir ncekinin uzerine eklenir ve model her admda iyiletirilerek istenen hedefe doru ilerler.

- En sk kullanlan boosting metodlar, AdaBoost (Adaptive Boosting), Gradient Boosting ve Stochastic Gradient Boosting metodlardr.
    - AdaBoost: hatal tahmin edilen gzlemlerin arlklandrarak hatay dzeltmeye alr.
    - Gradient Boosting: iteratif olarak hatay tahmin eder
    - Stochastic Gradient Boosting: her admda veri kmesinden rneklem seerek ilerler

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, classification_report, accuracy_score
```

```python
df = pd.read_csv("./diabetes.csv")
df.head()
```

```python
X = df.drop(["Outcome"], axis=1)
y = df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
preds_rf = rf.predict(X_test)
```

```python
print("Scikit-Learn's Random Forest Classifier's prediction accuracy is: %3.2f" % (accuracy_score(y_test, preds_rf)*100))
```

```python
import itertools

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues, fig=""):

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalize edilmi karmaklk matrisi")
    else:
        print('Normalize edilmemi karmaklk matrisi')
    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.grid(False)
    #plt.savefig(fig + ".png")
    plt.show()
```

```python
cm_rf = confusion_matrix(y_test, preds_rf)
plot_confusion_matrix(cm_rf, classes=['0', '1'], title='Confusion Matrix - Random Forest', fig="confusion_matrix_rf")
```

```python
print("....................................")
print(classification_report(y_test, preds_rf))
```
