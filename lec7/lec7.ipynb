**Karar Ağaçları**

- Karar ağaçları, verileri sınıflandırmaya veya girdilerden beklenen çıktıları tahmin etmeye imkan veren akış diyagramlarına benzeyen yapılardır.

- Özniteliklerin değerlerine göre yapılan ayrımlar (dallar) birbirine benzeyen örnekleri ayrıştırmaya dayanır. Bu dallar kendi içerisinde tutarlı yani benzer olmalıdır. Veriler dallara ayrılırken kural setleri oluşturuludur. Daha sonra bu kural setlerine göre sınıflandırma veya regresyon işlemi gerçekleştirilir.

Avantajları:
- Görselleştirmek ve uygulamak oldukça kolaydır. 
- Hiyerarşik yapısı ilişkilerin yönünü ve önemini ayırt etme kolaylığı sağlar.
- Hesaplama maliyeti düşüktür.
- Varsayımları kısıtlıdır.
- Kullanılan modelin yorumlanması bir dereceye kadar kolaydır.
- Kayıp ve eksik değerler karar ağaçların kullanılmasına engel değildir.

Dezavantajları:
- Çok sınıflı problemlerde başarısı düşüktür.
- Eğitim verisinin az olduğu durumlarda başarı oranı düşüktür.
- Ağaç budama karmaşıklığa neden olabilir.
- Düşük yanlılık (bias), yüksek varyans (variance)
- Veriye ve alınan kararlara aşırı duyarlı

### Topluluk Öğrenme
- Makine öğrenimi algoritmaları normalde tek bir sınıflandırıcı kullanarak tahminler yapmaya çalışırken topluluk öğrenme algoritmaları birden çok sınıflandırıcı kullanarak tahminler yapmaya çalışır.

**Bagging (Yeniden örneklem yöntemi veya Torbalama)**

- Leo Breiman tarafından önerilmiştir (1996). Amaç, sınıflandırma ve regresyon problemlerinde doğruluğunu arttırmak için geliştirilmiş bir tekniktir.

- Bagging (Bootstrapping Aggregating) metodu, var olan bir eğitim setinden yeni eğitim setleri türeterek temel öğreniciyi yeniden eğiten bir yöntemdir. Her bir veri seti için eğitilen temel öğrenicinin belirli bir test örneği üzerindeki kararları ortalama alınarak hesaplanır. 
- Bagging’de amaç yeni veri setleri türeterek farklılıkları oluşturmak ve bu sayede toplam sınıflandırma başarısını artırmaktır. Bu yöntemde eğitim setinden yaklaşık olarak %63,2 kadar orijinal örnek **rastgele** alınır ve alınan örneklerden bazıları çoğaltılarak (resampling) eğitim seti %100’e tamamlanır. 
- Bu yöntemle birbirinden farklı bir miktar eğitim seti elde edilir. Her eğitim seti aynı temel öğreniciye uygulanır ve alınan kararlar ağırlıklı oylama yöntemiyle birleştirilir (Sınıflandırma için). 
- Regresyon problemlerinde ise temel öğrenicilerin kararları ortalama ile belirlenir.

Ayrıca, 
- Yüksek varyanslı olan değişkenleri kullanarak onları daha uygun hale getirir. 
- Bagging yöntemi, tekil ağaçlara göre daha etkin sonuçlar vermektedir.
- Alt örneklemlerin her birinde bir model (zayıf model) oluşturulmuştur.
- Modeller paralel olarak çalışır ve birbirinden bağımsızdır.

Özetle iki temel üzerine kuruludur
- Bootstrap ile yeni örneklemler oluşturmak 
- Rastgele Örneklem Seçimi

**Random Forest**

- Topluluk sınıflandırma yöntemlerinin çalışma ilkesi, bir sınıflandırıcının yerine birden çok sınıflandırıcının ürettiği sonuçlar kullanılarak, yeni gelen bir verinin sınıflandırılmaya çalışılmasıdır. Topluluk sınıflandırma algoritmaları içerisinde en yaygın kullanılanlardan birisi olan Rastgele Orman (Random Forest), Breiman ve Cutler tarafından önerilmiştir. Tekil olarak oluşturulan birçok karar ağacı bir araya getirilerek karar ormanı oluşturulur. Oluşturulan her karar ağacı veri setinden booststrap yöntemi ile birbirinden ayrı örneklemlerle seçilmektedir. Karar ağaçlarındaki tahminler çeşitli yöntemlerle (ortalama alma vb.) bir araya getirilerek son tahmin çıktısı hesaplanır.

- Random forest algoritmasında, kullanıcının belirlediği ağaç sayısı N ve her düğümde kullanılması düşünülen özellik sayısına m diyelim.
- Her alt örneklemde eğitim verisi InBag ve Outofbag diye ikiye ayrılır.
- InBag, eğitim verisinin (2/3)'ünü kapsar. Outofbag ise eğitim verisinin (1/3)'ünü. Outofbag verisi, sistemi test etmek için kullanılır. Bu sayede her bir alt örneklemde karar ağacı modeli oluşturulur. Her bir düğüm noktasında m özellikleri, tüm özellikler arasından rastgele olacak şekilde seçilir ve seçilen özellikler arasından en iyi dal bulunmaya çalışılır.

- Amaç, ağaçlar arasındaki korelasyonu azaltmakdır. Düşük korelasyon için rassal çeşitlilik çok önemli.

- Rastgele Ormanlar algoritması çok geniş bir uygulama alanı olmasıyla sığ makine öğrenme görevleri düşünüldüğünde en iyi ikinci algoritma olmuştur. Rastgele Ormanlar Kaggle isimli meşhur makine öğrenmesi yarışma platformunun hayata geçirildiği 2010 yılından, 2014 yılına gradyan hızlandırma makinelerinin ortaya çıkışına kadar en popüler algoritmaydı.

Avantajları,
- Overfitting durumuna karşı dayanıklıdır.
- Ağaç sayısı için kısıt yoktur.
- Budamaya gerek yoktur.
- Eksik veri problemine karşı dayanıklıdır.

**Boosting**

- Boosting, bir çok zayıf öğreniciyi bir araya getirerek bir güçlü öğrenici (strong learner) oluşturmak anlamına gelir. Bir çok boosting metodunun ana fikri, tahmin edicileri ardışık olarak eğitmektir. 

- Boosting yapısı ise bağımsız paralel yapılar yerine, sırayla birbirinin üzerine eklenen yapıya sahiptir. Bu tahmin yapma yöntemleri, bir önceki tahminlerin hatalarından öğrenilen mantığı geliştirerek sonuca daha çok yaklaşmayı hedefleyen modellerdir. Bu modellerde her seferinde bir öncekinin hatalarını göz önünde bulundurarak ilerlediği için paralel çalışan modellere göre daha hızlı çalışırlar. Boosting yöntemi, bagging yönteminden farklı olarak paralel karar ağaçlarının birleştirilmesi yerine her üretilen karar ağacı, sırayla hatayı minimize edecek şekilde çalışır. Temelde, bu tip modellerde her yeni karar ağacı belirli ağırlık oranlarında bir öncekinin üzerine eklenir ve model her adımda iyileştirilerek istenen hedefe doğru ilerler.

- En sık kullanılan boosting metodları, AdaBoost (Adaptive Boosting), Gradient Boosting ve Stochastic Gradient Boosting metodlarıdır.
    - AdaBoost: hatalı tahmin edilen gözlemlerin ağırlıklandırarak hatayı düzeltmeye çalışır.
    - Gradient Boosting: iteratif olarak hatayı tahmin eder
    - Stochastic Gradient Boosting: her adımda veri kümesinden örneklem seçerek ilerler

**Bagging**
- Ortalama (voting)
- Paralel
- Rassal örneklem / öznitelikler
- Rassal örneklem (korrelasyon azaltma)
- Aşırı uyuma dayanıklı
- Rassal öğrenme
- Variance (varyans) azaltma

**Boosting** 
- İteratif 
- Ağırlıklı ortalama 
- Değiştirilmiş veri kümesi
- Bütün veri kümesi
- Aşırı uyum olabilir
- Yavaş öğrenme
- Bias (yanlılık) azaltma

**Gradient Boosting**

- Gradyan artırma (Gardient Boosting), hem regresyon hem de sınıflandırma problemleri için bir makine öğrenmesi tekniğidir. Gardient Boosting, güçlü bir öğrenici oluşturmak için birtakım zayıf öğrenicileri birleştirerek, yükseltme tekniğini kullanır. 

- Gradient Boosting yönteminde her iterasyonda bir önceki iterasyondaki hata tahmin edilmeye çalışılır.

- Hatadan öğrenmek gradyanın tersi yönde ilerlemeye denk gelir.

- Örneklem ve öznitelikler rassal olarak seçilerek de ilerlenebilir (Stochastic Gradient Boosting).

- XGBoost algoritması gradyan artırma modelidir ve boosting algoritmasıyla çalışmaktadır.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, classification_report, accuracy_score
```

```python
df = pd.read_csv("./diabetes.csv")
df.head()
```

```python
X = df.drop(["Outcome"], axis=1)
y = df["Outcome"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)
preds_rf = rf.predict(X_test)
```

```python
print("Scikit-Learn's Random Forest Classifier's prediction accuracy is: %3.2f" % (accuracy_score(y_test, preds_rf)*100))
```

```python
import itertools

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues, fig=""):

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalize edilmiş karmaşıklık matrisi")
    else:
        print('Normalize edilmemiş karmaşıklık matrisi')
    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.grid(False)
    #plt.savefig(fig + ".png")
    plt.show()
```

```python
cm_rf = confusion_matrix(y_test, preds_rf)
plot_confusion_matrix(cm_rf, classes=['0', '1'], title='Confusion Matrix - Random Forest', fig="confusion_matrix_rf")
```

```python
print("....................................")
print(classification_report(y_test, preds_rf))
```
