**Perceptron?**

- Tek katmanlı bir işlem ünitesidir. Sadece girdi ve çıktı katmalarından oluşmaktadır. 
- Çıktı üniteleri bütün girdi ünitelerine bağlanmaktadır ve her bağlantının bir ağırlığı vardır. En basit tek katmanlı sinir ağı modeli perceptron’dur. Temel olarak eğitilebilecek tek bir yapay sinir hücresinden oluşmaktadırlar. Perceptron kavramı ilk kez 1957 yılında Cornell Üniversitesi’nden psikolog Frank Rosenbatt tarafından ortaya atılmıştır (Rosenblatt, F. (1957). The perceptron, a perceiving and recognizing automaton Project Para. Cornell Aeronautical Laboratory.).

**Tek Katmanlı Perceptronların Genel Notasyonu**

$$ \sigma \Bigg( \sum_{i=1}^{n} x_i w_i + b \Bigg) = \sigma(\mathbf{x^T w} + b) = \hat{y} $$

                              
\begin{equation}
  \sigma(z)=\begin{cases}
    1, & z\leq0\\
    0, & z>0.
  \end{cases}
\end{equation}

$$ b = -\theta$$

- x: bağımsız değişken - girdi parametreleri
- w: ağırlık parametresi
- b: bias (yanlılık) değeri

- Perceptron modeli 1969 yılına gelindiğinde rafa kaldırılmaya başlandı. Marvin Minsky ve Seymour Papert yaptıkları çalışmada (Minsky, M. L., & Papert, S. A. (1969). Perceptrons: an introduction to computational geometry. MA: MIT Press, Cambridge.) tek katmanlı perceptron’ların, basit problemler için geçerli iken, problemler zorlaştıkça çözümden uzaklaştıklarını gösterdiler. 
    * Perceptronlar, **XOR problemini** çözemedi.
    * Nöronlar, çıkmaz sokak mı? İlk "AI Kışı" nın başlangıcı.
    
- Özetle, Minsky ve Papert, Perceptron'un sadece linear ayrılabilir örüntüleri sınıflandırabilirken , linear olmayan durumlarda çözümden uzaklaştığını göstermişlerdir.

### Derin Öğrenme Modeli

- Derin öğrenme, birbirini takip eden katmanlarda veriler işlenirken daha anlamlı gösterimler elde etmeye çalışan makine öğrenmesinin alt alanıdır.
- **Derin** ile kastedilen birbirini takip eden  gösterim katmanları ifade edilmektedir. Oluşturulan modeldeki katsayısı ilgili modelin derinliğini belirtmektedir. 
- Günümüzde real-world projelerde kullanılan derin öğrenme modelleri, binlerce katman içermektedir. Buna karşılık diğer makine öğrenme algoritmaları bir veya iki katmandan oluşmaktadır. Bu tarz öğrenme **sığ öğrenme** olarak adlandırılmaktadır.
- Derin öğrenme algoritmalarında öğrenme işlemi, sinir ağı olarak adlandırılan ve birbirini takip edenn katmanlar sayesinde gerçekleşmektedir. Katmanlar sinir ağlarına benzeltilmiş olsa da kullandığımız modeller insan beyninin tam olarak bir kopyası değildir. İnsan da öğrenme işleminin tam olarak açıklanamaması yani matematiksel olarak ifade edilememisi bunun en büyük sebebidir.
